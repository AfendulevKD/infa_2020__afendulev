```python
import pygame
from pygame.draw import *

# Params
# General
pygame.init()

# Screen
FPS = 1
screen_width = 1588
screen_size_multiplicator = 0.5

# Car
defoult_car_width = 400

# Var simplification
s_s_m = screen_size_multiplicator

def flip_houses(x1:int, dx1:int):
    '''
    Mirrors object relative to middle_x position of the screen
    x1 is x_coordinate of upper_left point of the object
    dx1 is the width of the object
    '''
    
    return screen_width-x1-dx1


def rotate_car(x3:int, dx3:int):
    '''
    Rotates object to match opposit direction
    x3 is the left_center coordinate of the object
    dx3 is width of the object
    '''
    
    return defoult_car_width-x3-dx3

def draw_auto(x2:int, y2:int, size:int, reverse:bool):
    '''
    Draws a car
    x2 is the left_center x_coordinate of the car
    y2 is the left_center y_coordinate of the car
    size is ratio between actual size and defoult size
    reverse shows direction in witch object must be drown
    '''
    
    draw_auto_body(x2, y2, size, reverse)
    draw_auto_wheels(x2, y2, size, reverse)
    draw_auto_windows(x2, y2, size, reverse)
    draw_auto_exhoust(x2, y2, size, reverse)
    
    
def draw_auto_body(x2:int, y2:int, size:int, reverse:bool):
    '''
    Draws a car_body
    x2 is the left_center x_coordinate of the car
    y2 is the left_center y_coordinate of the car
    size is ratio between actual size and defoult size
    reverse shows direction in witch object must be drown
    '''
    
    if reverse:
        rect(screen, (0, 204, 255), (int(x2 * s_s_m),int(y2 * s_s_m), int(400 * s_s_m * size), int(75 * s_s_m * size)))
        rect(screen, (0, 204, 255), (((x2+rotate_car(80,200)*size) * s_s_m),int((y2 - 50 * size) * s_s_m), int(200 * s_s_m * size), int(60 * s_s_m * size)))
    else:
        rect(screen, (0, 204, 255), (int(x2 * s_s_m), int(y2 * s_s_m), int(400 * s_s_m * size), int(75 * s_s_m * size)))
        rect(screen, (0, 204, 255), ((int((x2 + 80 * size) * s_s_m)),int((y2 - 50 * size) * s_s_m), int(200 * s_s_m * size), int(60 * s_s_m * size)))

    
def draw_auto_wheels(x2:int, y2:int, size:int, reverse:bool):
    '''
    Draws a car_wheels
    x2 is the left_center x_coordinate of the car
    y2 is the left_center y_coordinate of the car
    size is ratio between actual size and defoult size
    reverse shows direction in witch object must be drown
    '''
    
    if reverse:
        circle(screen, (0, 0, 0), (round((x2+rotate_car(80,0)*size) * s_s_m), round((y2+65*size) * s_s_m)), round(40 * s_s_m*size))
        circle(screen, (0, 0, 0), (round((x2+rotate_car(330,0)*size) * s_s_m), round((y2+65*size) * s_s_m)), round(40 * s_s_m*size))
    else:
        circle(screen, (0, 0, 0), (round(int((x2 + 80 * size) * s_s_m)), round((y2+65*size) * s_s_m)), round(40 * s_s_m*size))
        circle(screen, (0, 0, 0), (round((x2+330*size) * s_s_m), round((y2+65*size) * s_s_m)), round(40 * s_s_m*size))
    
def draw_auto_windows(x2:int, y2:int, size:int, reverse:bool):
    '''
    Draws a car_windows
    x2 is the left_center x_coordinate of the car
    y2 is the left_center y_coordinate of the car
    size is ratio between actual size and defoult size
    reverse shows direction in witch object must be drown
    '''
    
    if reverse:
        rect(screen, (213, 246, 255), ((x2+rotate_car(180,70)*size) * s_s_m,(y2-40*size) * s_s_m, 70 * s_s_m*size, 40 * s_s_m*size))
        rect(screen, (213, 246, 255), ((x2+rotate_car(100,70)*size) * s_s_m,(y2-40*size) * s_s_m, 70 * s_s_m*size, 40 * s_s_m*size))
    else:
        rect(screen, (213, 246, 255), ((x2+180*size) * s_s_m,(y2-40*size) * s_s_m, 70 * s_s_m*size, 40 * s_s_m*size))
        rect(screen, (213, 246, 255), ((x2+100*size) * s_s_m,(y2-40*size) * s_s_m, 70 * s_s_m*size, 40 * s_s_m*size))


def draw_auto_exhoust(x2:int, y2:int, size:int, reverse:bool):
    '''
    Draws a car_exhoust
    x2 is the left_center x_coordinate of the car
    y2 is the left_center y_coordinate of the car
    size is ratio between actual size and defoult size
    reverse shows direction in witch object must be drown
    '''
    
    if reverse:
        rect(screen, (0, 0, 0), ((x2+rotate_car(-20,20)*size) * s_s_m,(y2+50*size) * s_s_m, 20 * s_s_m*size, 10 * s_s_m*size))
    else:
        rect(screen, (0, 0, 0), ((x2 - 20 * size) * s_s_m,(y2+50*size) * s_s_m, 20 * s_s_m*size, 10 * s_s_m*size))

# Initiate screen
screen = pygame.display.set_mode((round(794 * 2 * s_s_m), round(1123 * s_s_m)))    

# Draw background
rect(screen, (183, 196, 200), (0 * s_s_m, 0 * s_s_m, 2 * 794 * s_s_m, 1123 * s_s_m))
rect(screen, (83, 108, 103), (0 * s_s_m,690 * s_s_m, 2 * 794 * s_s_m, 433 * s_s_m))

# Draw houses
rect(screen, (219, 227, 226), (594 * s_s_m,20 * s_s_m, 180 * s_s_m, 770 * s_s_m))
rect(screen, (147, 167, 172), (20 * s_s_m,20 * s_s_m, 180 * s_s_m, 770 * s_s_m))
rect(screen, (147, 172, 167), (230 * s_s_m,60 * s_s_m, 180 * s_s_m, 730 * s_s_m))
rect(screen, (183, 200, 196), (150 * s_s_m,160 * s_s_m, 180 * s_s_m, 730 * s_s_m))
rect(screen, (107, 130, 130), (500 * s_s_m,120 * s_s_m, 180 * s_s_m, 770 * s_s_m))

# Draw mirrored houses
rect(screen, (219, 227, 226), (flip_houses(594,180) * s_s_m,20 * s_s_m, 180 * s_s_m, 770 * s_s_m))
rect(screen, (147, 167, 172), (flip_houses(20,180) * s_s_m,20 * s_s_m, 180 * s_s_m, 770 * s_s_m))
rect(screen, (147, 172, 167), (flip_houses(230,180) * s_s_m,60 * s_s_m, 180 * s_s_m, 730 * s_s_m))
rect(screen, (183, 200, 196), (flip_houses(150,180) * s_s_m,160 * s_s_m, 180 * s_s_m, 730 * s_s_m))
rect(screen, (107, 130, 130), (flip_houses(500,180) * s_s_m,120 * s_s_m, 180 * s_s_m, 770 * s_s_m))

# Draw cars
draw_auto(1400,900,0.5, True)
draw_auto(100,900,0.5, True)
draw_auto(1000,900,0.5, True)
draw_auto(650,900,0.5, True)
draw_auto(272,1000,1, False)
draw_auto(1000,1000,1, False)
draw_auto(-300,1000,1, False)

pygame.display.update()
clock = pygame.time.Clock()
finished = False

while not finished:
    clock.tick(FPS)
    
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            finished = True

pygame.quit()
```
